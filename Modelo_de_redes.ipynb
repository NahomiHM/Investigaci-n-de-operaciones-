## Nodos / Vertices
 Elemento fundamental en la representación de redes y sistemas, se utiliza para diversos propósitos en la Investigación de Operaciones
 La terminologia usada en programacion para referirnos  los nodos va a ser **nodes**

###*Ejemplo 1*

````python
G.add_node(1) #Este codigo agrega un nodo a la vez

G.add_nodes_from([2, 3, 4, 5]) # Este codigo lo que hace es añadir nodos  a una lista


print("Nodos del grafo:")
print(f"Número de nodos: {G.number_of_nodes()}")   #Dentro de las llaves nos ayuda a saber cuantos nodos hay en total
print(f"Lista de nodos: {list(G.nodes())}")         # G.nodos nos ayudara a saber cuales son estos nodos  , para que  nos devuleva en forma de lista llamamos a list
print(f"Nodos con atributos: {list(G.nodes(data=True))}")  #
````

## Aristas / Arcos

Se define como una conexión que une dos nodos.
En programacion se le conocera o lo agregaremos con la palabra **edge**
Si  se desea saber de donde a donde va cada linea se puede usar el codigo:


```
# nx.edges(G/va como guardamos el grafico)
```




###*Ejemplo 2.1*

````python
G.add_edge(1, 2)  # Añadimos  una sola arista

G.add_edges_from([(2, 3), (3, 4), (4, 5), (5, 1)])  #Múltiples aristas desde una lista

print(" Aristas del grafo:")
print(f"Número de aristas: {G.number_of_edges()}") #llamamos a la funcion para que nos de el numero d earistas
print(f"Lista de aristas: {list(G.edges())}")
print(f"Aristas con atributos: {list(G.edges(data=True))}")
nx.edges(G)   #Nos dice de donde a donde va cada linea
````

###*Ejemplo 2.2*
Ahora repitamos el ejercicio pero con letras

````python
G.add_edge("A", "B")  # Añadimos  una sola arista

G.add_edges_from([("C", "D"), ("F", "G"), ("H", "I"), ("J", "K")])  #Múltiples aristas desde una lista

G.add_edges_from([ ("Ñ", "O")]) #Agregamos una aarista extra


print(" Aristas del grafo:")
print(f"Número de aristas: {G.number_of_edges()}") #llamamos a la funcion para que nos de el numero de aristas , al no agregar al principio del codigo un G = nx.Graph () se agregaron las aristas del ejercicio anterior
print(f"Lista de aristas: {list(G.edges())}")
print(f"Aristas con atributos: {list(G.edges(data=True))}")

nx.edges(G)   #Nos dice de donde a donde va cada linea
````

Para elimnar nodos o aristas es similar

```

```




````python
# Nodo
G.remove_node(5)
print(f"Nodos después de eliminar nodo 5: {list(G.nodes())}")  #Nos dara como resultados toda una lista de los nodos de los dos ejercicios sin el nodo que le indicamos eliminar

G.remove_node("I")
print(f"Nodos después de eliminar nodo I: {list(G.nodes())}")
#Arista
G.remove_edge(1, 2)
print(f"Aristas después de eliminar arista (1,6): {list(G.edges())}")  #Eliminara las arista

G.remove_edge("A", "B")
print(f"Aristas después de eliminar arista (A,B): {list(G.edges())}")
````

````python
import networkx as nx #Iniciaremos como si estuvieramos en un nuevo colab
G = nx.DiGraph() #agregamos un grafico dirigido

#Empezamos colocando los arcos

G.add_edges_from ([("A", "C"), ("B", "C"), ("B", "D"), ("C", "D"), ("A", "B")]) #arcos  en forma de lista

nx.draw_networkx(G) #Dibujamos el grafo

print("Grafo inicial:")
print("Nodos:", list(G.nodes()))
print("Aristas:", list(G.edges()))
````

###*Ejemplo 3.2*
Dentro de esta seccion a ese mismo grafo le agregaremos los atributos para que se visualice de mejor manera , vamos usar el codigo



```
# nx.draw_networkx (G , aqui se pondra todas las caracteristicas que deseamos que tenga )
```
Dentro de este codigo vamos agregar las especificaciones , si lo deseamos de color azul los nodos , mas grande la aletra , si deseamos que dibuje las fechas o no.

````python
import networkx as nx #Iniciaremos como si estubieramos en un nuevo colab
G = nx.DiGraph() #agregamos un grefico dirigido

G.add_edges_from ([("A", "C"), ("B", "C"), ("B", "D"), ("C", "D"), ("A", "B")]) #arcos  en forma de lista

pos = nx.spring_layout(G, seed=42)  #Definimos cuantas veces queremos que se repita sin moverse
nx.draw_networkx(G ,  pos,  #  Grafo y posición a dibujar
                 with_labels=True,      #  Mostrar nombres de nodos (A, B, C, D)
                 node_color='pink',     # Color rosado para los nodos
                 node_size=950,         # Tamaño grande de los nodos (800 puntos)
                 font_size=12,          # Tamaño de letra para las etiquetas
                 font_weight='bold',    # Texto en negrita para mejor visibilidad
                 arrows=True,           # Dibujar flechas (por ser grafo dirigido)
                 arrowsize=20)          # Tamaño de las flechas

print("Grafo inicial:")
print("Nodos:", list(G.nodes()))
print("Aristas:", list(G.edges()))
````

````python
import networkx as nx
import matplotlib.pyplot as plt

# Configuración para Colab
%matplotlib inline

# Crear grafo dirigido
G = nx.DiGraph()

# Empezamos colocando los arcos
G.add_edges_from([("A", "C"), ("B", "C"), ("B", "D"), ("C", "D"), ("A", "B")])

# POSICIÓN FIJA: Usar seed en spring_layout
pos = nx.spring_layout(G, seed=42)  # seed=42 para posición reproducible, siempre fija

# Dibujar con posición fija
plt.figure(figsize=(8, 6))
nx.draw_networkx(G, pos)

plt.title("Grafo con Posición Fija")
plt.axis('off')
plt.tight_layout()
plt.show()

print("Grafo inicial:")
print("Nodos:", list(G.nodes()))
print("Aristas:", list(G.edges()))
````

###*Ejemplo 4.1*

````python
import networkx as nx
import matplotlib.pyplot as plt

# Configuración para Colab
%matplotlib inline
# Crear grafo no dirigido
G = nx.Graph()

# Añadir múltiples nodos
G.add_nodes_from(["A", "B", "C", "D", "E"])

# Añadir múltiples aristas
G.add_edges_from([("A", "B"), ("B", "C"), ("C", "D"), ("D", "E"), ("E", "A")])

print("Grafo no dirigido con múltiples elementos:")
print(f"Nodos: {list(G.nodes())}")
print(f"Aristas: {list(G.edges())}")

# Visualizar
plt.figure(figsize=(8, 6))
nx.draw_networkx(G, with_labels=True, node_color='lightgreen', node_size=500)
plt.title("Grafo No Dirigido ", fontweight='bold')
plt.axis('off')
plt.show()
````

````python
import networkx as nx
import matplotlib.pyplot as plt

# Configuración para Colab
%matplotlib inline

# Crear grafo no dirigido
G = nx.Graph()

# Añadir múltiples nodos
G.add_nodes_from(["A", "B", "C", "D", "E"])

# Añadir múltiples aristas
G.add_edges_from([("A", "B"), ("B", "C"), ("C", "D"), ("D", "E"), ("E", "A")])

print("Grafo no dirigido con múltiples elementos:")
print(f"Nodos: {list(G.nodes())}")
print(f"Aristas: {list(G.edges())}")

# POSICIÓN FIJA: Usar seed en spring_layout
pos = nx.spring_layout(G, seed=42)  # seed=42 para posición reproducible, siempre fija

# Visualizar con posición fija
plt.figure(figsize=(8, 6))
nx.draw_networkx(G, pos, with_labels=True, node_color='lightblue', node_size=500)
plt.title("Grafo No Dirigido con Posición Fija", fontweight='bold')
plt.axis('off')
plt.show()
````
