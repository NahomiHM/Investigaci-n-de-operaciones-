## *Arbol de expansion minima*
se refiere a la forma de vincular todos los nodos de una red utilizando la longitud mínima total de las ramas de conexión
Para resolver este tipo de problemas se necesita saber alguna terminologia la cual se vera adelante y despues se realizara un ejercicio de esta.
Para saber cual es el arboo de expansion minima se usara este nuevo codigo:


```
# miniman = nx.minima_spanning_tree(G, weight='weight')

```
para que nos arroje el numero total del arbol de expansion se usara:


```
# expansion_total = minima.size(weight='weigth')
```
estos dos son las que ayudaran para cuando se pida el arbol de expansion minima.





Ejemplo
 Una compañía de televisión que busca proporcionar servicios de cable a nuevas áreas residenciales para determinar la red de menor costo
Solucion:


````python
import networkx as nx
import matplotlib.pyplot as plt

# Creación de Grafos
# Grafo no dirigido (G) y dirigido (D, no usado pero definido)
G = nx.Graph()
D = nx.DiGraph()

# 2. Agregar Nodos
G.add_node("1")
G.add_nodes_from(["2", "3", "4", "5", "6"])

#Agregar Aristas Ponderadas , reecordando que entre comillas se  pone las aristas ,posteriormente se coloca la cantidad
# Definimos la lista de tuplas (nodo_u, nodo_v, peso)
aristas_ponderadas = [
    ("1", "2", 1),
    ("1", "3", 5),
    ("1", "4", 7),
    ("1", "5", 6),
    ("2", "5", 3),
    ("2", "3", 6),
    ("2", "4", 4),
    ("3", "6", 10),
    ("3", "4", 5),
    ("4", "6", 3),
    ("5", "4", 8)
]
G.add_weighted_edges_from(aristas_ponderadas) #añade las conexione sque s enecesita con su pesos respectivamente
````

empecemos colocando los codigos que son nuevos y que nos ayudara a facilitar el problema o los problemas de arbol de expansion minima

````python
# La función devuelve un nuevo grafo
minima = nx.minimum_spanning_tree(G, weight='weight') #esta funcion nos ayuda a saber el minimo arbol de expansion que hay

EXPANSION_TOTAL = minima.size(weight='weight') #minima s ele conoce al arbol de espansion , esta funcion nos ayudara a calcular el costo total directamente

print("-Resultado del Árbol de Expansión Mínima-")
print(f"Costo Total Mínimo de la Red: {EXPANSION_TOTAL} unidades.")
print("Aristas del minima seleccionadas:")
for u, v, data in sorted(list(minima.edges(data=True))):
    print(f"  ({u} - {v}): {data['weight']}")

#dibujamos el grafo
plt.figure(figsize=(12, 7))

# posicion de los nodos
pos = nx.spring_layout(G, seed=42)

#grafo original
nx.draw(G, pos, with_labels=True, node_color='lightblue',
        node_size=1500, font_weight='bold', edge_color='gray')

# peso de cada arista que le vamos  a dar
labels_all = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels_all, font_color='gray')

# vamos a dibujarlos con especificaciones distintas como se vio anteiormente en la tabla que hicimos en el tutorial
nx.draw_networkx_edges(G, pos, edgelist=minima.edges(), edge_color='pink', width=3)

plt.title(f"Grafo Original  (Costo: {EXPANSION_TOTAL})")
plt.show()
````

Empecemos resolviendo el problema...


````python
import networkx as nx
import matplotlib.pyplot as plt

#Creamos un grafo dirigido
G_dirigido = nx.DiGraph()

# agregamos las arista y las distancias que son los pesos
distancias_red = [
    # (origen, destino, distancia)
    ("1", "2", 100),
    ("1", "3", 30),
    ("3", "4", 10),
    ("4", "2", 15),
    ("3", "5", 60),
    ("4", "5", 50),
]
G_dirigido.add_weighted_edges_from(distancias_red)
````

empecemos con los codigos que se vieron en el texto anterior , vamos ir colocandolos y viendo que hacen

````python
# definición del Origen punto de partida conocido
origen = "1"

# rutas minimas a todos los nodos
todas_las_distancias = nx.shortest_path_length(G_dirigido, source=origen, weight='weight')
todas_las_rutas = nx.shortest_path(G_dirigido, source=origen, weight='weight')

#verificamos que si sean los que calculamos en clase
print("rutas minimas ")
print(f"Origen {origen}\n")

print("[Distancia Mínima (Etiqueta Permanente) y Ruta Óptima a Cada Nodo Alcanzable]")
````

empezamos con la lista y lo importante o las nuevas funciones que se agregan y nos facilitra el calculo de la ruta minima

````python
#lista
lista = set()
#ciclo
for nodo_destino, distancia in sorted(todas_las_distancias.items()):
    if nodo_destino != origen:
        ruta = todas_las_rutas[nodo_destino]

        # ver los resultados
        print(f"  A Ciudad {nodo_destino}: {distancia} millas")
        print(f"    Ruta: {' -> '.join(ruta)}")

        # añadir las aristas
        path_edges = list(zip(ruta, ruta[1:]))
        lista.update(path_edges)

plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G_dirigido, seed=42)
````

para dibujar esta ruta vmos hacer como lo haciamos anteriormente , agregando cosas nuevas

````python
# Dibujar el grafo dirigido
nx.draw(G_dirigido, pos, with_labels=True, node_color='skyblue',
        node_size=1500, font_weight='bold', edge_color='gray', arrows=True)

labels_all = nx.get_edge_attributes(G_dirigido, 'weight')
nx.draw_networkx_edge_labels(G_dirigido, pos, edge_labels=labels_all, font_color='black')

# Resaltar las aristas que forman parte de CUALQUIERA de las rutas mínimas encontradas
nx.draw_networkx_edges(G_dirigido, pos, edgelist=list(lista ), edge_color='pink', width=1.5, arrows=True)

plt.title(f"Rutas minima empezando {origen} a todos los Nodos")
plt.show()
````

````python
import networkx as nx
import matplotlib.pyplot as plt

# grafo dirigido
D_seervada = nx.DiGraph()

#aristas y distancias como se vio en ejemplo anterior
#ejempo hecho en clase
capacidades_seervada = [
    # (Origen, Destino, Capacidad)
    ("S", "A", 5),
    ("S", "B", 8),
    ("S", "C", 4),

    ("A", "D", 5),

    ("B", "D", 6),
    ("B", "E", 4),

    ("C", "E", 3),

    ("D", "T", 8),
    ("E", "T", 5),

    ("E", "F", 3),
    ("F", "T", 5),
]

# agregamos las aristas usando el atributo 'capacity' , como se puede observar en la tabla de arriba donde esta las terminologias
D_seervada.add_weighted_edges_from(capacidades_seervada, weight='capacity')

#empezamos con la fuente y el suminero
fuente = "S"
sumidero = "T"

# flujo maximo
flujo_max_val, flujo_dict = nx.maximum_flow(D_seervada, fuente, sumidero)

print("Flujo Maximo SEERVADA PARK ")
print(f"Fuente (S): {fuente}, Sumidero (T): {sumidero}")
print(f"\nFlujo Máximo Total de la Red: {flujo_max_val} unidades.")
````

es importante verificar porque en este tipo de flujo maximo , nos podemos perder facilmente

````python
# el calculo es igual al de clase
if flujo_max_val == 14:
    print(" resultado óptimo coincide.")
else:
    print(f" El Flujo Máximo del ejemplo no coincide con el cálculo ({flujo_max_val}).")


print("\nDistribución Óptima de Flujo (Flujo / Capacidad):")
flujo_etiquetas = {}
for u, v, data in D_seervada.edges(data=True):
    flujo = flujo_dict[u][v]
    capacidad = data['capacity']
    if flujo > 0:
        print(f"  Arco ({u} -> {v}): {flujo} / {capacidad}")
    flujo_etiquetas[(u, v)] = f'{flujo}/{capacidad}'
````

hacemos el dibujo para visualizar si s eparece al ejercicio que se hizo en clase

````python
#Grafo dibujado
plt.figure(figsize=(10, 7))
pos = nx.spring_layout(D_seervada, seed=42)

node_colors = ['gray' if n == fuente else 'pink' if n == sumidero else 'skyblue' for n in D_seervada.nodes()]
nx.draw(D_seervada, pos, with_labels=True, node_color=node_colors,
        node_size=1500, font_weight='bold', arrows=True)

nx.draw_networkx_edge_labels(D_seervada, pos, edge_labels=flujo_etiquetas, font_color='black', font_size=10)

plt.title(f"Flujo Máximo: S a T | Total: {flujo_max_val} unidades")
plt.show()
````
